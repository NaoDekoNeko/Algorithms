```python
import networkx as nx
import matplotlib.pyplot as plt

grafo = [
    [0, 10, 15, 20, 25, 30],
    [10, 0, 35, 25, 30, 15],
    [15, 35, 0, 30, 5, 20],
    [20, 25, 30, 0, 15, 35],
    [25, 30, 5, 15, 0, 20],
    [30, 15, 20, 35, 20, 0]
]

G = nx.Graph()

for i in range(len(grafo)):
    for j in range(i+1,len(grafo[i])):
        G.add_edge(i,j,weight=grafo[i][j])

pos = nx.spring_layout(G, k = 6)
nx.draw(G,pos,node_color='lightgreen',with_labels=True)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos=pos,edge_labels=labels)

plt.show()
```

```python
import igraph as ig

grafo = [[0,1,15,6],[2,0,7,3],[9,6,0,12],[10,4,8,0]]
G = ig.Graph(directed=True)

G.add_vertices(len(grafo))
G.vs["label"] = [str(i) for i in range(len(grafo))]

for i in range(len(grafo)):
    for j in range(len(grafo[i])):
        if grafo[i][j] != 0:
            G.add_edge(i,j,weight=grafo[i][j])

edge_labels = [e['weight'] for e in G.es]

layout = G.layout("kk")
visual_style = {}
visual_style["vertex_size"] = 30
visual_style["vertex_color"] = "white"
visual_style["vertex_label"] = G.vs["label"]
visual_style["edge_label"] = edge_labels
visual_style["layout"] = layout
visual_style["bbox"] = (600, 600)
visual_style["margin"] = 40

path = [0,1,3,2,0]
for i in range(len(path)-1):
    eid = G.get_eid(path[i], path[i+1])
    G.es[eid]["color"] = "red"

ig.plot(G,**visual_style)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

grafo = [
    [0, 10, 15, 20, 25, 30],
    [10, 0, 35, 25, 30, 15],
    [15, 35, 0, 30, 5, 20],
    [20, 25, 30, 0, 15, 35],
    [25, 30, 5, 15, 0, 20],
    [30, 15, 20, 35, 20, 0]
]
G = nx.DiGraph()

for i in range(len(grafo)):
    for j in range(len(grafo[i])):
        if grafo[i][j] != 0:
            G.add_edge(i,j,weight=grafo[i][j])

pos = nx.spring_layout(G,k=6)

edge_labels = nx.get_edge_attributes(G,'weight')
edge_colors = ['black' for edge in G.edges]

path = [0 ,1 ,5 ,2 ,4 ,3 ,0]
for i in range(len(path)-1):
    edge1 = (path[i], path[i+1])
    edge2 = (path[i+1], path[i])
    if edge1 in G.edges:
        edge_colors[list(G.edges).index(edge1)] = 'red'
    elif edge2 in G.edges:
        edge_colors[list(G.edges).index(edge2)] = 'red'

nx.draw(G,pos=pos)
nx.draw_networkx_edge_labels(G,pos=pos,edge_labels=edge_labels)
nx.draw_networkx_edges(G,pos=pos,edge_color=edge_colors)
nx.draw_networkx_labels(G,pos=pos)

plt.show()
```
